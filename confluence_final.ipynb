```python

import os
import pyperclip
import time
import pyautogui
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
 
 
chrome_options = Options()
navegador = webdriver.Chrome(options=chrome_options)
navegador.get("https://tigrendene.atlassian.net/wiki/spaces/TOTVS/overview?homepageId=134939227")
time.sleep(12)
 
# Login no Confluence
navegador.find_element(By.XPATH, '//*[@id="username"]').send_keys("luiz.nobre@grendene.com.br")
time.sleep(1)
navegador.find_element(By.XPATH, '//*[@id="login-submit"]/span').click()
time.sleep(5)
navegador.find_element(By.XPATH, '//*[@id="password"]').send_keys("R@drigues2003")  
time.sleep(5)
navegador.find_element(By.XPATH, '//*[@id="login-submit"]/span').click()
time.sleep(20)
navegador.find_element(By.XPATH,'//*[@id="216498315"]').click()
time.sleep(5)
 
# Caminho da pasta principal
base_folder = r'C:\Users\lz6321\OneDrive - grendene.com.br\MIGRAÇÃO_2\Financeiro _ Controladoria\Faturamento'
 # Copia o caminho para a área de transferência
 
 
 
if not os.path.exists(base_folder):
    print(f"O diretório {base_folder} não existe.")
else:
    try:
        # Selecionar a pasta no Confluence
        xpath_elemento = f"//span[contains(text(), 'Faturamento')]"
        elemento = navegador.find_element(By.XPATH, xpath_elemento)
        navegador.execute_script("arguments[0].scrollIntoView();", elemento)  # Garantir visibilidade
        elemento.click()
        time.sleep(5)
 
        # Processar todos os arquivos da pasta
        for file in os.listdir(base_folder):
            file_path = os.path.join(base_folder, file)
 
            # Verifica se é um arquivo .docx
            if os.path.isfile(file_path) and file.endswith(".docx"):  
                print(f"Importando arquivo: {file_path}")
 
                try:
                    # Criar nova subpágina para cada arquivo
                    xpath_botao = f"//button[.//span[contains(text(), 'Criar conteúdo filho para Faturamento')]]"
                    print(f"Procurando o botão: {xpath_botao}")
                    botao = WebDriverWait(navegador, 10).until(
                        EC.element_to_be_clickable((By.XPATH, xpath_botao))
                    )
                    print("Botão encontrado e clicável.")
                    botao.click()
                    time.sleep(5)
 
                    # Clicar na opção de criar uma página
                    xpath_pagina = "//a[.//span[text()='Página']]"
                    pagina_elemento = WebDriverWait(navegador, 10).until(
                        EC.element_to_be_clickable((By.XPATH, xpath_pagina))
                    )
                    pagina_elemento.click()
                    time.sleep(5)
 
                    try:
                        # Acessar a área de importação
                        botao_importar = WebDriverWait(navegador, 10).until(
                            EC.element_to_be_clickable((By.XPATH, '//*[@id="content-body"]/div/div/div/div[1]/div[2]/div[2]/div[3]/div[4]/div/div/div/button'))
                        )
                        botao_importar.click()
                        time.sleep(5)
 
                        # Selecionar "Templates e importação de documentos"
                        xpath_criar = "//button[.//span[text()='Templates e importação de documentos']]"
                        criar_elemento = WebDriverWait(navegador, 10).until(
                            EC.element_to_be_clickable((By.XPATH, xpath_criar))
                        )
                        criar_elemento.click()
                        time.sleep(5)
 
                        # Acessar a aba de importação
                        aba_importacao = WebDriverWait(navegador, 10).until(
                            EC.element_to_be_clickable((By.XPATH, '//*[@id="editorContextPanelContentCreationTabs-1"]/span/span'))
                        )
                        aba_importacao.click()
                        time.sleep(5)
 
                        # Selecionar arquivo
                        botao_selecionar_arquivo = WebDriverWait(navegador, 10).until(
                            EC.element_to_be_clickable((By.XPATH, '//*[@id="editorContextPanelContentCreationTabs-1-tab"]/div/div[1]/button/div[2]/span[1]/div/span'))
                        )
                        botao_selecionar_arquivo.click()
                        time.sleep(5)
 
                    except Exception as e:
                        print(f"Erro ao acessar a área de importação para o arquivo {file}: {e}")
                        continue  # Pula para o próximo arquivo se der erro
 
                    # Copia o caminho para a área de transferência
                    pyperclip.copy(file_path)
 
                    # Cola o caminho no campo de upload
                    pyautogui.hotkey("ctrl", "v")
                    time.sleep(5)
 
                    # Pressiona "Enter" para confirmar
                    pyautogui.press("enter")
                    time.sleep(10)
 
                    # Finalizar importação e publicar a página
                    navegador.find_element(By.XPATH, '//*[@id="com-atlassian-confluence"]/div[2]/div[7]/div/div[2]/div/div/section/div[2]/button/span/span').click()
                    time.sleep(5)
                    navegador.find_element(By.XPATH, '//*[@id="content-body"]/div/div/div/div[1]/div[2]/div[2]/div[3]/div[1]/div[1]/div/button/span').click()
                    time.sleep(5)
                    navegador.find_element(By.XPATH, '//*[@id="publish-button"]/span/span').click()
                    time.sleep(5)
                    navegador.get("https://tigrendene.atlassian.net/wiki/spaces/TOTVS/overview?homepageId=134939227")
                    time.sleep(12)
                    navegador.find_element(By.XPATH,'//*[@id="216498315"]').click()
                    time.sleep(5)
                    # Selecionar a pasta no Confluence
                    xpath_elemento = f"//span[contains(text(), 'Faturamento')]"
                    elemento = navegador.find_element(By.XPATH, xpath_elemento)
                    navegador.execute_script("arguments[0].scrollIntoView();", elemento)  # Garantir visibilidade
                    elemento.click()
                    time.sleep(5)
 
 
                    print(f"Página criada para o arquivo: {file}")
 
                except Exception as e:
                    print(f"Erro ao processar o arquivo {file}: {e}")
 
    except Exception as e:
        print(f"Erro ao processar a pasta {base_folder}: {e}")
 
print("Páginas criadas com sucesso no Confluence.")
 
 

 
```
