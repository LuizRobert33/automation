```python
#Código com atualização
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
import os
from selenium import webdriver as opcoesSelenium
from selenium.webdriver.common.keys import Keys
import pyautogui as tempoEspera
import pyautogui as teclasAtalhoTeclado
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os


nome_arquivo_bens = "C:\\Users\\lz6321\\Downloads\\Mov_Bem_Sintetico_G_Info.xlsx"
planilhaDadosBens = load_workbook(nome_arquivo_bens)
sheet_selecionada = planilhaDadosBens["Movimento Bem Sintético"]


navegador = opcoesSelenium.Chrome()
navegador.get("https://grendene.agidesk.com/br/portal/login?redirect=painel")



tempoEspera.sleep(8)
navegador.find_element(By.NAME, "username").send_keys("luiz.nobre@grendene.com.br") 
tempoEspera.sleep(8)
navegador.find_element(By.NAME, "password").send_keys("R@drigues2003") 
tempoEspera.sleep(10)
navegador.find_element(By.NAME, "password").send_keys(Keys.RETURN)

tempoEspera.sleep(20)
navegador.find_element(By.XPATH, '//*[@id="sidebar-app-menu-40"]').click()


def descontinuado():
    tempoEspera.sleep(15)
    return navegador.find_element(By.XPATH, "//tr[starts-with(@id, 'asset-row-')]//td[4]//span[@title='Descontinuado']")

def desativado():
    tempoEspera.sleep(7)
    desativado_elemento = navegador.find_element(By.XPATH, "//tr[starts-with(@id, 'asset-row-')]//td[4]//span[@title='Desativado']")
    tempoEspera.sleep(5)
    elemento = navegador.find_element(By.XPATH, "//tr[starts-with(@id, 'asset-row-')]//td[6]//a")
    elemento.click()

def ativo():
    tempoEspera.sleep(7)
    return navegador.find_element(By.XPATH, "//tr[starts-with(@id, 'asset-row-')]//td[4]//span[@title='Ativo']")

def status():
    tempoEspera.sleep(5)
    navegador.find_element(By.XPATH, '//*[@id="assets-main-form-fields-10-field-0"]/div/input').send_keys("Descontinuado")
    tempoEspera.sleep(3)
    navegador.find_element(By.XPATH, '//*[@id="assets-main-form-fields-10-field-0"]/div/div[2]/div').click()

def salvar():
    tempoEspera.sleep(5)
    navegador.find_element(By.XPATH, "//button[starts-with(@id, 'button-') and @title='Salvar']").click()
    tempoEspera.sleep(15)
    navegador.find_element(By.XPATH, "//button[starts-with(@id, 'button-') and @title='Voltar']").click()
    tempoEspera.sleep(5)

def conta(linha):
    tempoEspera.sleep(5)
    conta_valor = sheet_selecionada['C%s' % linha].value
    navegador.find_element(By.XPATH, '//*[@id="1039"]').send_keys(conta_valor)

def planilha_success(linha):
    sheet_selecionada['F%s' % linha].fill = fill_success

def planilha_failure(linha):
    sheet_selecionada['F%s' % linha].fill = fill_failure

def npatrimonio(BensPesquisa):
    tempoEspera.sleep(5)
    navegador.find_element(By.NAME, "filter").send_keys(BensPesquisa)
    tempoEspera.sleep(5)
    navegador.find_element(By.NAME, "filter").send_keys(Keys.RETURN)

def apagar():
    tempoEspera.sleep(10)
    navegador.find_element(By.XPATH, '//*[@id="assets-erase-link"]').click()

def patrimonio():
    tempoEspera.sleep(7)
    navegador.find_element(By.XPATH, '//*[@id="simple-assets-form-action-field-2"]/div/input').send_keys("n")
    tempoEspera.sleep(6)
    navegador.find_element(By.XPATH, '//*[@id="simple-assets-form-action-field-2"]/div/div[2]/div[11]').click()

def status_inicial():
    tempoEspera.sleep(7)
    navegador.find_element(By.XPATH, '//*[@id="simple-assets-form-fields-0-field-0"]/div/input').send_keys("desativado")
    tempoEspera.sleep(6)
    navegador.find_element(By.XPATH, '//*[@id="simple-assets-form-fields-0-field-0"]/div/div[2]').click()


fill_success = PatternFill(start_color="92D050", end_color="92D050", fill_type="solid")
fill_failure = PatternFill(start_color="FF0000", end_color="FF0000", fill_type="solid")

valores_processados = set()

for linha in range(12, len(sheet_selecionada['D']) + 1):
    apagar()
    patrimonio()
    tempoEspera.sleep(7)
    
    BensPesquisa = sheet_selecionada['D%s' % linha].value
    if not BensPesquisa:
        break
    
    if BensPesquisa in valores_processados:
        continue
    valores_processados.add(BensPesquisa)

    npatrimonio(BensPesquisa)

    try:
        descontinuado()
        planilha_success(linha)

    except NoSuchElementException:
        try:
            desativado()
            status()
            salvar()

            try:
                descontinuado()
                planilha_success(linha)

            except NoSuchElementException:
                try:
                    desativado()
                    status()
                    conta(linha)
                    salvar()
                    planilha_success(linha)

                except NoSuchElementException:
                    planilha_failure(linha)
                    

        except NoSuchElementException:
            planilha_failure(linha)
           

# Salvar e enviar planilha
tempoEspera.sleep(2)
planilhaDadosBens.save(filename=nome_arquivo_bens)
tempoEspera.sleep(2)
os.startfile(nome_arquivo_bens)

navegador.quit()

print('Leitura da planilha concluída!')
```
