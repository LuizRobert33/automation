```python
import json
import re

caminho_arquivo = "C:\\Users\\lz6321\\OneDrive - grendene.com.br\\Documentos\\automation\\response(viagem_total)"

def extrair_dados(content):
    data_ida = re.search(r'Data e hora de ida.*?(\d{2}/\d{2}/\d{4} \d{2}:\d{2})', content, re.IGNORECASE | re.DOTALL)
    data_retorno = re.search(r'Data e hora de retorno.*?(\d{2}/\d{2}/\d{4} \d{2}:\d{2})', content, re.IGNORECASE | re.DOTALL)
    nome = re.search(r'Nome do colaborador que fará a viagem \*?\s*\n([^\n]+)', content, re.IGNORECASE)
    justificativa = re.search(r'Objetivo e justificativa\s*\n([^\n]+)', content, re.IGNORECASE)
    centro_custo = re.search(r'^[ \t]*Centro de custo \*?\s*\n([^\n]+)', content, re.IGNORECASE | re.MULTILINE)
    local = re.search(r'Destino \*?\s*\n([^\n]+)', content, re.IGNORECASE)

    data_ida_val = data_ida.group(1) if data_ida else 'Não encontrado'
    data_retorno_val = data_retorno.group(1) if data_retorno else 'Não encontrado'
    nome_val = nome.group(1).strip() if nome else 'Não encontrado'
    justificativa_val = justificativa.group(1).strip() if justificativa else 'Não encontrado'
    centro_custo_val = centro_custo.group(1).strip() if centro_custo else 'Não encontrado'
    local_val = local.group(1).strip() if local else 'Não encontrado'

    return data_ida_val, data_retorno_val, nome_val, justificativa_val, centro_custo_val, local_val

with open(caminho_arquivo, encoding='utf-8') as f:
    issues = json.load(f)

for issue in issues:
    issue_id = issue.get('id', 'Sem ID')
    content = issue.get('content', '')
    data_ida, data_retorno, nome, justificativa, centro_custo, local = extrair_dados(content)
    print(f"ID: {issue_id}")
    print(f"Data de ida: {data_ida}")
    print(f"Data de retorno: {data_retorno}")
    print(f"Nome do colaborador: {nome}")
    print(f"Justificativa: {justificativa}")
    print(f"Centro de custo: {centro_custo}")
    print(f"Local: {local}")
    print('-' * 40)
```
